## Resetting the multiindex in order to sort by County and to create a date column

```python
by_month = pd.read_pickle('by_month.pk1')

reset_by_month = by_month.reset_index(['County', 'Year', 'Month'])
reset_by_month['date'] = pd.to_datetime(reset_by_month['Year'].astype(str) + '-' + reset_by_month['Month'].astype(str))
```

## Drought Plots

```python
Balt_by_month = reset_by_month[(reset_by_month['County'] == 'Baltimore City')]
Balt_by_month.plot('date', 'Weighted Drought')
```

```python
Bos_by_month = reset_by_month[(reset_by_month['County'] == 'Suffolk County')]
Bos_by_month.plot('date', 'Weighted Drought')
```

```python
Chi_by_month = reset_by_month[(reset_by_month['County'] == 'Cook County')]
Chi_by_month.plot('date', 'Weighted Drought')
```

```python
Nola_by_month = reset_by_month[(reset_by_month['County'] == 'Orleans Parish')]
Nola_by_month.plot('date', 'Weighted Drought')
```
